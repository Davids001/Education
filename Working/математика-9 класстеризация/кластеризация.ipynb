```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('snsdata.csv')
data.head()
```

```python
data.info()
```

```python
df = data.iloc[:, 4:]
display(df.head())
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
col = df.columns
X_norm = scaler.transform(df)
X_norm = pd.DataFrame(X_norm,columns=col)
X_norm
```

```python
X = df.values
scaler = StandardScaler()
scaler.fit(X)
X_norm = scaler.transform(X)
X_norm[0][0]
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=9, random_state=42)
model.fit(X_norm)
model.inertia_
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=9,random_state=42)
model.fit(X_norm)
model.inertia_
```

```python
X_norm['pred_y'] = model.predict(X_norm)
X_norm['pred_y'].value_counts()
```
