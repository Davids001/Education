```python
import pandas as pd
solar_data = pd.read_csv('solarpower.csv')
solar_data.head()
```

```python
for i in range(len(solar_data['cum_power'])):
    if solar_data['cum_power'][i] ==  -1:
        solar_data['cum_power'][i] = (solar_data['cum_power'][i-1] + solar_data['cum_power'][i+1])/2
for i in solar_data['cum_power']:
    if i == -1:
        print('a') 
        
        
           
```

```python
round(solar_data['cum_power'].mean(),)
```

```python
solar_data['day_power'] = solar_data['cum_power'].diff()
solar_data
```

```python
solar_data['date'] = pd.to_datetime(solar_data['date']).dt.to_period('M')
```

```python
solar_data
```

```python
zen = solar_data.groupby('date',as_index=False)['day_power'].mean()
zen
```

```python
import plotly
import plotly.express as px
fig =px.line(
    zen,
    x='date',
    y = 'day_power',
    );
fig.show()
```
